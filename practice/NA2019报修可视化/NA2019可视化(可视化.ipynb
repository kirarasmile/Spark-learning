```python
# 1.每日的报修单数->柱状图
def drawChart_1(index):
    root = "/home/hadoop/smile/NA/NA1/part-00000-ec9138e4-1329-4085-86df-c92106154f44-c000.json"
    date = []
    count = []
    with open(root, 'r') as f:
        while True:
            line = f.readline()
            if not line:
                break
            js = json.loads(line)
            date.append(str(js['date']))
            count.append(str(js['date_order']))
    d = (
    Bar()
        .add_xaxis(date)
        .add_yaxis("2019年每日单数", count, stack="stack1", color="#B0C4DE")
        .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
        .set_global_opts(title_opts=opts.TitleOpts(title="每日单数"))
        .render("/home/hadoop/smile/NA/NA1/NA1.html")
    )
    return d
    
# drawChart_1(1).render_notebook()
drawChart_1(1)
```

```python
# 每栋楼的报修数->象柱状图
def drawChart_2(index):
    root = "/home/hadoop/smile/NA/NA2/part-00000-0261ad7a-7f49-463f-9fba-0cecf5f6e433-c000.json"
    dst = []
    total = []
    with open(root, 'r') as f:
        while True:
            line = f.readline()
            if not line:
                break
            js = json.loads(line)
            dst.insert(0, str(js['dst']))
            total.insert(0, int(js['cdst']))
        
    a = (
        PictorialBar()
            .add_xaxis(dst)
            .add_yaxis(
                "楼层报修数",
                total,
                label_opts = opts.LabelOpts(is_show = False),
                symbol_size = 18,
                symbol_repeat = "fixed",
                symbol_offset = [0, 0],
                is_symbol_clip = True,
                color = "#FFD700",
                symbol = SymbolType.ROUND_RECT,
            )
            .reversal_axis()
            .set_global_opts(
                title_opts = opts.TitleOpts(title = "各楼报修一览"),
                xaxis_opts = opts.AxisOpts(is_show = True),
                yaxis_opts = opts.AxisOpts(
                    axistick_opts = opts.AxisOpts(is_show = False),
                    axisline_opts = opts.AxisLineOpts(
                        linestyle_opts = opts.LineStyleOpts(opacity = 0)    
                    ),
                ),
            )
            .render("/home/hadoop/smile/NA/NA2/NA2.html")
    )
    
    return a

# drawChart_2(1).render_notebook()
drawChart_2(1)
```

```python
# 每个月报修单数->柱状图
def drawChart_4(index):
    root = "/home/hadoop/smile/NA/NA4/part-00000-d23bf302-25ae-435b-a69c-a0b8ec02488b-c000.json"
    month = []
    count = []
    with open(root, 'r') as f:
        while True:
            line = f.readline()
            if not line:
                break
            js = json.loads(line)
            month.append(str(js['month']))
            count.append(str(js['count']))
    c = (
        EffectScatter()
        .add_xaxis(month)
        .add_yaxis("", count)
        .set_global_opts(
            title_opts=opts.TitleOpts(title="每月报修单数"),
            xaxis_opts=opts.AxisOpts(splitline_opts=opts.SplitLineOpts(is_show=True), name="月份"),
            yaxis_opts=opts.AxisOpts(splitline_opts=opts.SplitLineOpts(is_show=True)),
        )
        .render("/home/hadoop/smile/NA/NA4/NA4.html")
    )
    return c
    
# drawChart_4(1).render_notebook()
drawChart_4(1)
```

```python

def drawChart_5(index):
    root = "/home/hadoop/smile/NA/NA3/part-00000-f8d6da7c-ba6e-488f-8b71-ef58e69b521d-c000.json"
    values= []
    sum1 = 0
    sexcount = []
    with open(root, 'r') as f:
        while True:
            line = f.readline()
            if not line:
                break
            js = json.loads(line)
            sexcount.append(str(js['sexcount']))
        sum1 = float(sexcount[0]) + float(sexcount[-1])
        count1 = float(sexcount[0])/ float(sum1)
        values.append(["男(%)", round(float(count1)*100, 2)])
        values.append(["女(%)", 100-round(float(count1)*100, 2)])
                
    a = (
        Pie()
            .add("", values)
            .set_colors(["#6495ED", "pink"])
            .set_global_opts(title_opts = opts.TitleOpts(title = "男女比例"))
            .set_series_opts(label_opts = opts.LabelOpts(formatter = "{b}: {c}"))
            .render("/home/hadoop/smile/NA/NA3/NA3.html")
    )
    
    return a

# drawChart_5(8).render_notebook()
drawChart_5(8)
```
